```
#library list as well as models and dataset also the visualization tools. 
import seaborn as sns 
import numpy as np
from sklearn.svm import SVC 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import classification_report, confusion_matrix 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
```

# **Load the data**

```
import pandas as pd
#loading the dataset from seaborn
a=sns.load_dataset('penguins')   
a=a.dropna()
sns.countplot(a['species'])
```

```
#visualizing the data with pairplot to see data similarity
sns.pairplot(a, hue='island', palette='Dark2')
```

```
#visualizing the data with pairplot to see data similarity
sns.pairplot(a, hue='sex', palette='Dark2')
```

```
#visualizing the data with pairplot to see data similarity
sns.pairplot(a, hue='species', palette='Dark2')
```

# **SVM**

```
X=a.drop('species' , axis=1)
X=X.drop('island' , axis=1) 
X=X.drop('sex' , axis=1) 
y=a['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)  
model=SVC() 
clf=model.fit(X_train,y_train)   
print(cross_val_score(model,X, y, cv=5))
```

```
 predictsvm=clf.predict(X_test) 
predictsvm
```

# **KNN**

```
X=a.drop('species' , axis=1)
X=X.drop('island' , axis=1) 
X=X.drop('sex' , axis=1)
y=a['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)  
model=KNeighborsClassifier()
clf=model.fit(X_train,y_train)  
 
predictknn=clf.predict(X_test) 
predictknn
```

# Gaussian Naive **Bayes**

```
X=a.drop('species' , axis=1)
X=X.drop('island' , axis=1) 
X=X.drop('sex' , axis=1)
y=a['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)  
model=GaussianNB()
clf=model.fit(X_train,y_train)  
 
predictgnb=clf.predict(X_test) 
predictgnb
```

**Print both classification report from SVM and KNN to see comparison between two classifier **

```
print (classification_report(y_test, predictsvm))
print ('Report Of SVM') 
print (classification_report(y_test, predictknn))  
print ('Report Of KNN') 
print (classification_report(y_test, predictgnb))  
print ('Report Of GNB') 
print(confusion_matrix(y_test, predictsvm)) 
print('/n') 
print(confusion_matrix(y_test,predictknn)) 
print('/n') 
print(confusion_matrix(y_test,predictgnb))
```
